```python
#Drop rows with missing data in column 'daily vaccinations'

daily_vaccinations = data.dropna(subset=['daily_vaccinations'])

daily_vaccinations.isnull().sum()
```

```python
daily_vaccinations_countries = (
    daily_vaccinations[['country', 'daily_vaccinations']]
    .groupby('country')
    .mean()
    .rename(columns={'daily_vaccinations': 'daily_vaccinations_mean'})
    .sort_values(by='daily_vaccinations_mean', ascending=False)
    .reset_index('country')
    
)
daily_vaccinations_countries.head(5)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
#Prepare date about daily mean vaccinations by countries, to use in chart. Let's choose random 15 countries.

data_dmv = daily_vaccinations_countries.sample(15)
```

```python
plt.figure(figsize=(30,10))
plt.plot(data_dmv.country, data_dmv.daily_vaccinations_mean ,"m*-.",linewidth=0.5,markersize=10, label= 'Daily vaccinations mean by countries') 
plt.grid(False)
plt.xlabel('Country')
plt.ylabel('Daily Vaccinations Mean')
plt.legend(loc="lower left")
plt.savefig('mean_daily_vaccinations.pdf')
plt.show()
```

```python
#Create similiar chart as above, but use data with top 5 and bottom 5 vaccinating countries and compare it with Poland.

new_data = (
    daily_vaccinations_countries.head(5) 
    .append(daily_vaccinations_countries.tail(5), ignore_index=True)
    .append(daily_vaccinations_countries[daily_vaccinations_countries.country == 'Poland'], ignore_index= True)
    .sort_values(by='daily_vaccinations_mean', ascending= False)
)
new_data
```

```python
plt.figure(figsize=(25,10))
plt.plot(new_data.country, new_data.daily_vaccinations_mean ,"m*-.",linewidth=0.5,markersize=15, label= 'Daily vaccinations mean by countries') 
plt.grid(False)
plt.xlabel('Country')
plt.ylabel('Daily Vaccinations Mean')
plt.legend(loc="lower left")
plt.savefig('mdv_Poland_vs_world.pdf')
plt.show()
```

```python
# I would like to create correlation matrix for vaccines use in each country and the mean od daily vaccinations.
# First, I will prepare data to analyse types of vaccines in individual countries.

vaccines_data = data.dropna(subset=['daily_vaccinations'])[['country', 'daily_vaccinations', 'vaccines']]
vaccines_data
```

```python
# Then I will check types of vaccines used
vaccines = vaccines_data.vaccines.unique() 
all_vaccines = []
for n in list(vaccines):
    for vaccine in n.split(","):
        all_vaccines.append(vaccine)
all_vaccines = set(all_vaccines)
all_vaccines
```

```python
all_vaccines = list(all_vaccines)
for i in range(2):
    for vaccine in all_vaccines:
        if vaccine[0] == ' ':
            vaccine1 = vaccine[1:]
            all_vaccines.append(vaccine1)
            all_vaccines.remove(vaccine)
all_vaccines = set(all_vaccines)
all_vaccines
```

```python
#Finally we got all types of vaccines used. Now let's add to dataframe columns with individual vaccines. 
# If vaccine was used in country, the value will be 1. If not, 0.

for vaccine in all_vaccines:
    vaccines_data.loc[vaccines_data['vaccines'].str.contains(vaccine), vaccine] = 1
    vaccines_data[vaccine].fillna(0.0, inplace=True)
vaccines_data
```

```python
vaccines_data.drop(columns=['vaccines'], inplace=True)
vaccines_data
```

```python
vaccines_data = (
    vaccines_data
    .groupby(['country', 'Moderna', 'Oxford/AstraZeneca', 'Sputnik V', 'Pfizer/BioNTech', 'Johnson&Johnson','Sinovac','Sinopharm/Wuhan','Covaxin','EpiVacCorona', 'Sinopharm/Beijing'  ])
    [['daily_vaccinations']]
    .mean()
    .rename(columns={'daily_vaccinations': 'mean_daily_vaccinations'})
    .reset_index()
)
```

```python
#Create correlation matrix
corr_matrix = vaccines_data.iloc[:,0:12].corr()
plt.figure(figsize = (30,10))
sns.heatmap(corr_matrix, annot=True)
plt.show()
```

```python
# Compare charts of daily vaccinations in period of time for different countries and look for patterns between them.

#Select 5 countries
countries_list = ['Poland', 'United Kingdom', 'United States', 'Norway']
data_5 = data[data.country.isin(countries_list)]

#Drow rows where is missing data in 'daily_vaccinations' column
data_5 = data_5.dropna(subset=['daily_vaccinations'])
data_5['daily_vaccinations'].isnull().sum()
```

```python
data_5 = data_5.groupby(['date'])[['country']].sum()
```

```python
data_5.loc[data_5['country'].str.contains('NorwayPolandUnited KingdomUnited States'), 'data_required'] = 1
data_5['data_required'].fillna(0.0, inplace=True)
data_5
```

```python
selected_dates = data_5[data_5['data_required'] == 1].reset_index()['date']
len(selected_dates)
```
