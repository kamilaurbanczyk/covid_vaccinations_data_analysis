```python
#Drop rows with missing data in column 'daily vaccinations'

daily_vaccinations = data.dropna(subset=['daily_vaccinations'])

daily_vaccinations.isnull().sum()
```

```python
daily_vaccinations_countries = (
    daily_vaccinations[['country', 'daily_vaccinations']]
    .groupby('country')
    .mean()
    .rename(columns={'daily_vaccinations': 'daily_vaccinations_mean'})
    .sort_values(by='daily_vaccinations_mean', ascending=False)
    .reset_index('country')
    
)
daily_vaccinations_countries.head(5)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
#Prepare date about daily mean vaccinations by countries, to use in chart. Let's choose random 15 countries.

data_dmv = daily_vaccinations_countries.sample(15)
```

```python
plt.figure(figsize=(30,10))
plt.plot(data_dmv.country, data_dmv.daily_vaccinations_mean ,"m*-.",linewidth=0.5,markersize=5, label= 'Daily vaccinations mean by countries') 
plt.grid(False)
plt.xlabel('Country')
plt.ylabel('Daily Vaccinations Mean')
plt.legend(loc="lower left")
plt.savefig('mean_daily_vaccinations.pdf')
plt.show()
```
